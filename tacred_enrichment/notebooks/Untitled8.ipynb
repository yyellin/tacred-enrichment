```python
import numpy as np
import pickle


def load(filename):
    with open(filename, 'rb') as infile:
        id_to_embedding = pickle.load(infile)
        embedding_to_id = dict([(id_to_embedding[idx], idx) for idx in range(len(id_to_embedding))])
    return id_to_embedding, embedding_to_id
```
