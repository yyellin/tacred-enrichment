```python
import pandas

train_ud = pandas.read_csv(r'C:\Users\JYellin\re_1\tacred\results\general\train-ud_paths_with_ner_v0.0.6.csv')
```

```python
relations = train_ud['relation'].unique().tolist()
```

```python
stats = []
overall_total = 0
overall_matched = 0

for relation in relations:
    total = train_ud[(train_ud['relation']==relation)].shape[0]
    matched = train_ud[(train_ud['relation']==relation) & ((train_ud['type1']==train_ud['type1_corenlp']) & (train_ud['type2']==train_ud['type2_corenlp']))].shape[0]
    overall_total += total
    overall_matched += matched
    
    stats.append( (relation, round(matched/total,4)) )
    
stats = sorted(stats, key=lambda x: x[1], reverse=True)

stats.insert(0,('overall', round(overall_matched/overall_total,4)) )

    
```
