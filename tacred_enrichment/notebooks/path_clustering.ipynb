```python
import pandas
pandas.set_option('display.max_rows', 100)

ud_train = pandas.read_csv(r'C:\Users\jyellin\re_1\Code\path_to_re\train-ud_paths_per_relation.csv')
ud_train = ud_train[ud_train['relation'] == 'org:founded_by']

```

```python
path_to_count = ud_train.groupby(['path'])['id'].count().reset_index().rename(columns={'id': 'total'}).sort_values(['total'], ascending=[False]).reset_index(drop=True)
path_to_count
```

```python
#df.loc[df['Language'].apply(lambda x: x in ['C', 'C++'] )])
high_counts = path_to_count.loc[ path_to_count['path'].apply(lambda x: len(x.split())<4) ]
high_counts
```

```python
def path_to_grakel_graph(path):
    steps = path.split(' ')

    edges = []
    node_names = {}
    edge_names = {}
    
    node_id = 0
    node_names[node_id] = 'F'
    
    for step_i, step in enumerate(steps):
        
        node_name = 'E' if step_i < len(steps)-1 else 'F'
        node_id += 1
        
        node_names[node_id] = node_name
        
        direction = step[0]
        if direction == '>':
            edges.append( (node_id-1, node_id) )
            edge_names[(node_id-1, node_id)] = step[1:]
        else:
            edges.append( (node_id, node_id-1) )
            edge_names[(node_id, node_id-1)] = step[1:]
        
    
    return (edges, node_names, edge_names)
```

```python
unique_paths = high_counts['path'].to_list()
```

```python
X = []
for path in unique_paths:
    X.append(path_to_grakel_graph(path))
```

```python
from grakel import RandomWalk

wl_subtree = RandomWalk(normalize=False, method_type='baseline')
#wl_subtree = GraphKernel(kernel=['weisfeiler_lehman', 'edge_histogram'], normalize=False)


#wl_subtree = GraphKernel(kernel=['WL', 'ST-WL'], normalize=False)
#wl_subtree = GraphKernel(kernel=['neighborhood_subgraph_pairwise_distance'], normalize=False)



K = wl_subtree.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, verbose=0)
kmeans.fit(K)
```

```python
cluster_mapping = pandas.DataFrame()
cluster_mapping['path'] = unique_paths
cluster_mapping['cluster'] = kmeans.labels_
cluster_mapping.sort_values(['cluster'], ascending=[True])
```

```python
import matplotlib.pyplot as plt
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black')
```
